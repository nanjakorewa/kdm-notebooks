```python
import os
import pandas as pd
import seaborn as sns
from full_fred.fred import Fred

# FRED_API_KEY = os.getenv('FRED_API_KEY')
fred = Fred()
print(f"FRED APIキーが環境変数に設定されている：{fred.env_api_key_found()}")
```

## Unemployment Rate
ソース：https://fred.stlouisfed.org/series/UNRATE

```python
df_UNRATE = fred.get_series_df("UNRATE")[["date", "value"]].copy()
df_UNRATE["date"] = pd.to_datetime(df_UNRATE["date"])
df_UNRATE["value"] = pd.to_numeric(df_UNRATE["value"], errors="coerce")
df_UNRATE = df_UNRATE.set_index("date")
df_UNRATE = df_UNRATE.loc["2010-01-01":]
df_UNRATE.head(5)
```

```python
sns.set(rc={"figure.figsize": (15, 8)})
sns.lineplot(data=df_UNRATE, x="date", y="value")
```
